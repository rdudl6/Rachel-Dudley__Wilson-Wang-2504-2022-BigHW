```julia
#Q3 b
using Plots
function hail_length(x::Int)
    n = 0
    while x != 1
        x = ((x % 2 == 0) ? (x ÷ 2) : (3x + 1))
        n += 1
    end
    return n
end


function hail_max(N)
    max_n = -Inf
    for x in 1:N
        n = hail_length(x)
        if n > max_n
            max_n = n
        end
    end
    return max_n
end

step_range = 1:10^4
max_vals = [hail_max(N) for N in step_range]
plot(step_range, max_vals, xaxis=:log, 

    label=false, xlabel="Maximal initial value", ylabel="Maximal number of steps")
```
